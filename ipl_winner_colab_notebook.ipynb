```python
# Dataset config
DATASET = "saurav9786/indian-premier-league-match-analysis"
file_path = "matches.csv"  # Set to the file you want from the dataset
TARGET_COL = "winner"
RANDOM_STATE = 42

# Load dataset file as DataFrame
df = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    DATASET,
    file_path
)

print("Loaded file:", file_path)
print("Shape:", df.shape)
display(df.head())
```

```python
TEAM_ALIASES = {
    "Delhi Daredevils": "Delhi Capitals",
    "Kings XI Punjab": "Punjab Kings",
    "Rising Pune Supergiant": "Rising Pune Supergiants",
    "Pune Warriors": "Pune Warriors India",
}

LEAKAGE_COLS = {
    "winner",
    "result",
    "result_margin",
    "win_by_runs",
    "win_by_wickets",
    "player_of_match",
    "method",
    "umpire1",
    "umpire2",
    "umpire3",
}

def normalize_team_name(name):
    if pd.isna(name):
        return name
    return TEAM_ALIASES.get(name, name)

def normalize_team_columns(frame):
    frame = frame.copy()
    team_cols = [c for c in ["team1", "team2", "winner", "toss_winner"] if c in frame.columns]
    for c in team_cols:
        frame[c] = frame[c].map(normalize_team_name)
    return frame

def prepare_dataframe(frame):
    frame = frame.copy()
    frame.columns = [c.strip() for c in frame.columns]
    frame = normalize_team_columns(frame)

    if "date" in frame.columns:
        frame["date"] = pd.to_datetime(frame["date"], errors="coerce")
        frame["match_year"] = frame["date"].dt.year
        frame["match_month"] = frame["date"].dt.month
        frame["match_dayofweek"] = frame["date"].dt.dayofweek

    if "season" not in frame.columns and "match_year" in frame.columns:
        frame["season"] = frame["match_year"]

    frame = frame[frame[TARGET_COL].notna()].copy()

    if "team1" in frame.columns and "team2" in frame.columns:
        valid = (frame[TARGET_COL] == frame["team1"]) | (frame[TARGET_COL] == frame["team2"])
        frame = frame[valid].copy()

    return frame

def build_features(frame):
    use_cols = [c for c in frame.columns if c not in LEAKAGE_COLS]
    X = frame[use_cols].copy()
    y = frame[TARGET_COL].copy()

    miss_ratio = X.isna().mean()
    drop_cols = miss_ratio[miss_ratio > 0.6].index.tolist()
    X = X.drop(columns=drop_cols, errors="ignore")

    if "date" in X.columns:
        X = X.drop(columns=["date"])

    cat_cols = X.select_dtypes(include=["object", "category", "bool"]).columns.tolist()

    for c in cat_cols:
        X[c] = X[c].astype(str).replace("nan", "Unknown").fillna("Unknown")

    num_cols = [c for c in X.columns if c not in cat_cols]
    for c in num_cols:
        X[c] = pd.to_numeric(X[c], errors="coerce")
        X[c] = X[c].fillna(X[c].median())

    return X, y, cat_cols

def time_split(frame):
    if "season" in frame.columns:
        seasons = sorted(frame["season"].dropna().unique())
        if len(seasons) >= 2:
            valid_season = seasons[-1]
            train_idx = frame["season"] < valid_season
            valid_idx = frame["season"] == valid_season
            if train_idx.sum() > 100 and valid_idx.sum() > 20:
                return train_idx, valid_idx, f"season split: train < {valid_season}, valid = {valid_season}"

    if "date" in frame.columns and frame["date"].notna().sum() > 0:
        f2 = frame.sort_values("date")
        cutoff = int(len(f2) * 0.8)
        train_ids = set(f2.iloc[:cutoff].index.tolist())
        train_idx = frame.index.isin(train_ids)
        valid_idx = ~train_idx
        return train_idx, valid_idx, "date split: first 80% train, last 20% valid"

    cutoff = int(len(frame) * 0.8)
    train_idx = np.zeros(len(frame), dtype=bool)
    train_idx[:cutoff] = True
    valid_idx = ~train_idx
    return train_idx, valid_idx, "index split: first 80% train, last 20% valid"

print("Functions ready.")
```

```python
df_clean = prepare_dataframe(df)
X, y, cat_cols = build_features(df_clean)
train_idx, valid_idx, split_info = time_split(df_clean)

X_train = X.loc[train_idx].copy()
X_valid = X.loc[valid_idx].copy()
y_train = y.loc[train_idx].copy()
y_valid = y.loc[valid_idx].copy()

cat_features_idx = [X_train.columns.get_loc(c) for c in cat_cols if c in X_train.columns]

model = CatBoostClassifier(
    loss_function="MultiClass",
    eval_metric="TotalF1",
    iterations=1200,
    depth=8,
    learning_rate=0.03,
    l2_leaf_reg=5,
    random_seed=RANDOM_STATE,
    verbose=200
)

model.fit(
    X_train,
    y_train,
    cat_features=cat_features_idx,
    eval_set=(X_valid, y_valid),
    use_best_model=True
)

valid_probs = model.predict_proba(X_valid)
valid_pred = model.predict(X_valid).ravel()

acc = accuracy_score(y_valid, valid_pred)
ll = log_loss(y_valid, valid_probs, labels=model.classes_)

print("Split strategy:", split_info)
print("Train size:", len(X_train))
print("Valid size:", len(X_valid))
print(f"Validation Accuracy: {acc:.4f}")
print(f"Validation Log Loss: {ll:.4f}")
```

```python
print("Classification report:")
print(classification_report(y_valid, valid_pred, zero_division=0))
```

```python
# Save model artifacts
MODEL_PATH = "ipl_winner_catboost.joblib"
META_PATH = "ipl_model_meta.json"

joblib.dump(model, MODEL_PATH)
with open(META_PATH, "w", encoding="utf-8") as f:
    json.dump({
        "feature_cols": X.columns.tolist(),
        "cat_cols": cat_cols,
        "target_col": TARGET_COL,
        "created_at": datetime.utcnow().isoformat() + "Z",
        "dataset": DATASET,
        "file_path": file_path
    }, f, indent=2)

print("Saved:", MODEL_PATH, "and", META_PATH)
```

```python
def predict_match_winner(model, feature_cols, cat_cols, row_dict, top_k=3):
    row = pd.DataFrame([row_dict])

    for c in ["team1", "team2", "winner", "toss_winner"]:
        if c in row.columns:
            row[c] = row[c].map(normalize_team_name)

    if "date" in row.columns:
        row["date"] = pd.to_datetime(row["date"], errors="coerce")
        row["match_year"] = row["date"].dt.year
        row["match_month"] = row["date"].dt.month
        row["match_dayofweek"] = row["date"].dt.dayofweek
        row = row.drop(columns=["date"])

    for c in feature_cols:
        if c not in row.columns:
            row[c] = np.nan

    row = row[feature_cols].copy()

    for c in feature_cols:
        if c in cat_cols:
            row[c] = row[c].astype(str).replace("nan", "Unknown").fillna("Unknown")
        else:
            row[c] = pd.to_numeric(row[c], errors="coerce").fillna(0)

    probs = model.predict_proba(row)[0]
    classes = model.classes_
    ranked = sorted(zip(classes, probs), key=lambda x: x[1], reverse=True)[:top_k]
    return ranked

sample_input = {
    "season": 2024,
    "city": "Mumbai",
    "venue": "Wankhede Stadium",
    "team1": "Mumbai Indians",
    "team2": "Chennai Super Kings",
    "toss_winner": "Mumbai Indians",
    "toss_decision": "field",
    "dl_applied": 0
}

preds = predict_match_winner(model, X.columns.tolist(), cat_cols, sample_input, top_k=2)
print("Top predictions:")
for team, p in preds:
    print(f"{team}: {p:.3f}")
```

```python
# Interactive prediction form
import ipywidgets as widgets
from IPython.display import display, clear_output

teams = sorted(list(set(pd.concat([df_clean["team1"], df_clean["team2"]]).dropna().unique()))) if "team1" in df_clean.columns and "team2" in df_clean.columns else []
cities = sorted(df_clean["city"].dropna().unique().tolist()) if "city" in df_clean.columns else ["Unknown"]
venues = sorted(df_clean["venue"].dropna().unique().tolist()) if "venue" in df_clean.columns else ["Unknown"]
seasons = sorted(df_clean["season"].dropna().astype(int).unique().tolist()) if "season" in df_clean.columns else [2024]

w_season = widgets.Dropdown(options=seasons, description="Season:")
w_team1 = widgets.Dropdown(options=teams, description="Team 1:")
w_team2 = widgets.Dropdown(options=teams, description="Team 2:")
w_toss_winner = widgets.Dropdown(options=teams, description="Toss Win:")
w_toss_decision = widgets.Dropdown(options=["bat", "field"], description="Toss Dec:")
w_city = widgets.Dropdown(options=cities, description="City:")
w_venue = widgets.Dropdown(options=venues, description="Venue:")
w_dl = widgets.Dropdown(options=[0, 1], value=0, description="D/L:")

btn = widgets.Button(description="Predict Winner", button_style="success")
out = widgets.Output()

def on_predict(_):
    with out:
        clear_output()
        if w_team1.value == w_team2.value:
            print("Team 1 and Team 2 must be different.")
            return

        row = {
            "season": w_season.value,
            "city": w_city.value,
            "venue": w_venue.value,
            "team1": w_team1.value,
            "team2": w_team2.value,
            "toss_winner": w_toss_winner.value,
            "toss_decision": w_toss_decision.value,
            "dl_applied": w_dl.value
        }

        preds = predict_match_winner(model, X.columns.tolist(), cat_cols, row, top_k=3)
        print("Predicted win probabilities:")
        for team, p in preds:
            print(f"{team}: {p:.2%}")

btn.on_click(on_predict)
display(w_season, w_team1, w_team2, w_toss_winner, w_toss_decision, w_city, w_venue, w_dl, btn, out)
```
